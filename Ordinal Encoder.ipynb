```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('shop_data.csv')
df
```

```python
df['income'].unique()
```

```python
income_order = ['low', 'medium', 'high']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enco = OrdinalEncoder(categories=[income_order])
```

```python
enco.fit(df[['income']])
```

```python
final = pd.DataFrame(enco.transform(df[['income']]))
```

```python
df['income'] = final
```
