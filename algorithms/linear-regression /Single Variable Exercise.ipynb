```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/canada_per_capita_income.csv")
df.head(3)
```

```python
%matplotlib inline
plt.xlabel('Year)')
plt.ylabel('per capita income (US$)')
plt.scatter(df.year, df['per capita income (US$)'], c='red', marker='+')
```

```python
from sklearn import linear_model
```

```python
rg = linear_model.LinearRegression()
rg.fit(df[['year']], df['per capita income (US$)'])
```

```python
rg.predict([[2020]])
```

```python
rg.coef_
```

```python
rg.intercept_
```

```python
# Y = m * X + b (m is coefficient and b is intercept)
2020*828.46507522 + (-1632210.7578554575)
```

```python
%matplotlib inline
plt.xlabel('Year', fontsize=20)
plt.ylabel('Per capita income (US$)', fontsize=20)
plt.scatter(df.year, df['per capita income (US$)'], c='red', marker='+')
plt.plot(df.year, rg.predict(df[['per capita income (US$)']]), color='blue')
```
